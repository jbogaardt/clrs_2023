The strength between the connections between neurons must be fine-tuned during training. But how is it done? After the end of each prediction, the deviation of the prediction from the ground truth is measured. Through Stochastic Gradient Descent, we can determine how much of the total error can be attributed to the output signal of each neuron in the network. This process is called "backpropagation". Figure 5 shows a simple neural network with only 2 neurons, one acting as the gradient and the other acting as the intercept of an OLS regression.

Figure 6 shows data created out of a linear regression process with parameters of B0 = -0.75 and B1 = 1.25

```python
def make_dataset(weights_array, losses, B0, B1):
    b1_hat = [x[0].flatten()[0] for x in weights_array]
    b0_hat = [x[1].flatten()[0] for x in weights_array]
    b1 = [B1]*len(b1_hat)
    b0 = [B0]*len(b0_hat)
    combined_order = [*range(len(b0))]
    combined_b1 = b1_hat+b1
    combined_label = ['Estimate']*len(b1_hat) + ['Truth']*len(b1)
    combined_b0 = b0_hat+b0
    df = pd.DataFrame()
    df['B0'] = combined_b0
    df['B1'] = combined_b1
    df['Step'] = combined_order*2
    df['Type'] = combined_label
    df['Losses'] = losses*2
    return df

rand_mat = np.random.randn(200,1,1).astype(np.float32)
B0, B1 = -0.75, 1.25

X_train = rand_mat[:150]
y_train = (B0 + B1*X_train) + 0.75*np.random.standard_normal(150).astype(np.float32).reshape(-1,1,1)
X_test = rand_mat[150:]
y_test = (B0 + B1*X_test) + 0.75*np.random.standard_normal(50).astype(np.float32).reshape(-1,1,1)

mae = tf.keras.losses.MeanAbsoluteError()
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0175)
model = tf.keras.models.Sequential([
            tf.keras.Input(shape=(1,1)),
            #tf.keras.layers.Dense(2, activation='relu'),
            tf.keras.layers.Dense(1, activation='elu')
        ])

epochs = 175
network_internals = []
losses = []
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        output_value = model(X_train, training=True)
        loss_value = mae(y_train, output_value)
        losses.append(np.sum(loss_value))
    grads = tape.gradient(loss_value, model.trainable_weights)
    network_internals.append(model.get_weights().copy())
    optimizer.apply_gradients(zip(grads, model.trainable_weights))

fig = px.scatter(x=X_train.flatten(), y=y_train.flatten())
fig.show()
```

As can be seen, the training process repeatedly optimizes the output values of neurons over the traning loop, making the network progressively predict better results. The last slide of the animation shows the final values of B0 and B1.

```python
df = make_dataset(network_internals, losses, B0, B1)
fig = px.scatter(df, x="B1", y="B0", animation_frame="Step", color="Type", hover_data=['Losses'])
fig.update_xaxes(range=[-2,2])
fig.update_yaxes(range=[-1,1])
fig.show()
```

## Classes & Functions

ProcessData - A class which consumes a ChainLadder object and contains the following methods:
- processing_triangles - Selects a line of business
- decompose_triangles - Extract Loss Rectangles for each firm, in terms of link ratios
- train_test_split - Split data into training and testing datasets.

model_creation - Creates various types of Deep Neural Networks.

```python
class ProcessData:

    def __init__(self, triangle_object):
        self.triangle_object = triangle_object
        self.train_x, self.train_y = None, None
        self.val_x, self.val_y = None, None
        self.test_x, self.test_y = None, None
        self.lob = None
        self.no_lags = None
        self.company_names = None
        self.payment_type = None
        self.no_of_firms = None
        self.firm_names = None
        self.model_input_shape = None
        self.data_dict = {}

    def processing_triangles(self, lob, payment_type):
        self.lob = lob
        self.payment_type = payment_type
        self.triangle_object = self.triangle_object[self.triangle_object["LOB"] == lob][payment_type]
        self.no_of_firms = self.triangle_object.shape[0]
        self.company_names = self.triangle_object['GRNAME']

    def decompose_triangles(self, number_of_accident_yrs=10):
        for count, name in enumerate(self.company_names):
            temp = cl.Chainladder().fit(self.triangle_object.iloc[count])
            # temp = temp.full_triangle_.link_ratio.loc[..., '12-24':'108-120'].to_frame().to_numpy()
            temp = temp.full_triangle_.link_ratio.to_frame().iloc[:number_of_accident_yrs,
                   :number_of_accident_yrs - 1].to_numpy()
            if not np.isnan(temp).any():
                self.data_dict[name] = temp.copy()

    def train_test_split(self, train_prop, val_prop, no_of_lags=3, lstm=True):
        data = np.array(list(self.data_dict.values()), dtype=np.float32)
        length = len(data)
        lags = data.shape[-1]
        x = data[:, :, :lags - no_of_lags]
        y = data[:, :, lags - no_of_lags:]
        if lstm:
            x = tf.transpose(x, [0, 2, 1])
            y = tf.transpose(y, [0, 2, 1])
        self.model_input_shape = x.shape[1:]
        self.no_lags = no_of_lags
        self.train_x, self.val_x, self.test_x = x[:int(length * train_prop)], \
            x[int(length * train_prop):int(length * (train_prop + val_prop))], x[int(length * (train_prop + val_prop)):]
        self.train_y, self.val_y, self.test_y = y[:int(length * train_prop)], \
            y[int(length * train_prop):int(length * (train_prop + val_prop))], y[int(length * (train_prop + val_prop)):]
        self.train_x, self.val_x, self.test_x = tf.convert_to_tensor(self.train_x, dtype=tf.float32), \
            tf.convert_to_tensor(self.val_x, dtype=tf.float32), tf.convert_to_tensor(self.test_x, dtype=tf.float32)
        self.train_y, self.val_y, self.test_y = tf.convert_to_tensor(self.train_y, dtype=tf.float32), \
            tf.convert_to_tensor(self.val_y, dtype=tf.float32), tf.convert_to_tensor(self.test_y, dtype=tf.float32)


def model_creation(input_shape, out_len, lstm=True):
    if lstm:
        model = tf.keras.models.Sequential([
            tf.keras.Input(shape=input_shape),
            tf.keras.layers.LSTM(256, activation='elu', return_sequences=False),
            #tf.keras.layers.LSTM(64, activation='elu', return_sequences=False),
            tf.keras.layers.Dense(out_len*input_shape[-1]),
            tf.keras.layers.Reshape((out_len, input_shape[-1]))
        ])
    else:
        model = tf.keras.models.Sequential([
            tf.keras.Input(shape=input_shape),
            tf.keras.layers.Dense(256, activation='elu'),
            tf.keras.layers.Dense(256, activation='elu'),
            #tf.keras.layers.Dense(80, activation='relu'),
            tf.keras.layers.Dense(out_len)
        ])
    model.summary()
    return model
```

## Feed Forward Neural Networks (FFNN)

The example of the Neural Network that we saw before is a very simple one, although it was able to approximate the parameters of the OLS regression fairly accurately. For data with possible non-linear patterns, we need to look at more complex architectures. This neural network will be several orders of magnitude more complex than the previous one and hence difficult to show as before.
Our FFNN only has 2 hidden layers and 1 output layer. Each layer only contains 64 neurons. Please keep in mind that this is only one possible architecture and that there are a myriad of other ways to build one. For our loss function, we have selected the Mean Absolute Percentage Error (MAPE) function, provided in Figure 10.

mape.svg
<center>Figure 10</center>
Figure 10 : https://en.wikipedia.org/wiki/Mean_absolute_percentage_error

ffnn_flow.png
<center>Figure 11</center>

As shown above in Figure 11, the Feed Forward Neural Network model takes in the first loss development lag for all accident years, per company to predict the the lags from 2 to 9 for all accident years. Therefore the output prediction is in a matrix form. Since we are providing a 3D input in the format of : (No. of companies, Accident Years, Lag Periods) and because our hidden layers have 64 neurons each, we are using essentially using 64 neurons to calculate the output sequence for each accident year.

```python
model_ffn = model_creation(data_object_ffn.model_input_shape, data_object_ffn.no_lags, lstm=False)
model_ffn.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                  loss=tf.keras.losses.MeanAbsolutePercentageError(
                       #reduction=tf.keras.losses.Reduction.NONE,
                       name='mean_absolute_percentage_error'
                        ),
                  metrics='mean_absolute_percentage_error')
model_ffn.fit(data_object_ffn.train_x,
              data_object_ffn.train_y,
              batch_size=1,
              validation_split=0.2,
              epochs=150)
prediction_ffn = model_ffn.predict(data_object_ffn.test_x)
```

```python
performance_ffn = np.round(
    (tf.math.reduce_mean(
        tf.math.cumprod(
            (prediction_ffn/data_object_ffn.test_y),
            axis=-1),
        axis=0)
     -1)*100,
    2)
x = ['LDF 2', 'LDF 3', 'LDF 4', 'LDF 5', 'LDF 6', 'LDF 7', 'LDF 8', 'LDF 9']
y = ['AY 1', 'AY 2', 'AY 3', 'AY 4', 'AY 5', 'AY 6', 'AY 7', 'AY 8', 'AY 9', 'AY 10']
fig = px.imshow(performance_ffn,
                x = x,
                y = y,
                text_auto=True, width=600, height=800)
fig.update_layout(
    title='Percentage Deviance of the Predictions, at each Lag Period')
fig.show()
```

## Long-Short Term Memory (LSTM) Network

For LSTM networks, instead of using plain neurons to process data, we use a more sophisticated form of cells within the network. It is quite easy for us to see that in our prediction problem, we have to predict a squence, not a single value, for each of the Accident Years. LSTMs are quite good at memorizing long sequence since they have 2 forms of memory, Long and Short. Long term memory is held in the cell state (C(t-1)) whilst the short term memory is held in the hidden state (H(t-1)). More details : https://en.wikipedia.org/wiki/Long_short-term_memory

LSTM_Cell.png
<center>Figure 12</center>

Figure 12 : By Guillaume Chevalier - File:The_LSTM_Cell.svg, CC BY-SA 4.0, https://commons.wikimedia.org/w/index.php?curid=109362147

The LSTM based model has 2 additional input layers which are made of LSTM cells, a single hidden dense layer with another output layer.

lstm_flow.png
<center>Figure 13</center>

When dealing with LSTMs, we have to be carful to format data in a manner which is acceptable for LSTMs. LSTMs accept 3D data but they must be in the following order : (No. of squences, Length of each feature, No. of features). Therefore in our case, we can think of each accident year as a feature; and since for the 'X' we only provide the first lag, the length of each respective feature is 1. Our no. of squences is our number of firms. Conversely, what we are predicting out of the model should also be in the previous format. This means that we can't feed data as is and must change the axis of data for LSTMs. This is shown in Figure 13.

## Tuner

There are many variables that we can fine tune in a network. One of the most straightforward variables to fine tune is the number of neurons in each hidden layer and the activation function of each hidden layer. The package Keras-Tuner lets us run tests on various randomized combinations of these two types of variables. Let's see how much more optimized we can make the LSTM network.

```python
data_object_lstm.train_test_split(0.6, 0.2, no_of_lags=8, lstm=True)

def tuned_lstm(hp):
    model = tf.keras.Sequential(
        [
            tf.keras.Input(shape=data_object_lstm.model_input_shape),
            tf.keras.layers.LSTM(units=hp.Int("units", min_value=128, max_value=756, step=64),
                                 activation=hp.Choice("activation", ["relu", "elu"]),
                                 return_sequences=False),
            tf.keras.layers.Dense(data_object_lstm.no_lags*data_object_lstm.model_input_shape[-1]),
            tf.keras.layers.Reshape((data_object_lstm.no_lags, data_object_lstm.model_input_shape[-1]))
        ]
    )
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss=tf.keras.losses.MeanAbsolutePercentageError(
                #reduction=tf.keras.losses.Reduction.NONE,
                name='mean_absolute_percentage_error'
            ),
        metrics='mean_absolute_percentage_error'
    )
    return model
```

```python
tuner = keras_tuner.RandomSearch(
    hypermodel=tuned_lstm,
    objective=keras_tuner.Objective("val_mean_absolute_percentage_error", direction="min"),
    max_trials=8,
    overwrite=True,
    directory="my_dir",
    project_name="built_in_metrics",
)

tuner.search(
    x=data_object_lstm.train_x,
    y=data_object_lstm.train_y,
    epochs=150,
    validation_data=(data_object_lstm.val_x, data_object_lstm.val_y),
)
```
